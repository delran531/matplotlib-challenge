```python
# Dependencies and Setup
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df = pd.read_csv(mouse_drug_data_to_load)
trial_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
complete_trial_df = pd.merge(mouse_df, trial_df, how="left", on=["Mouse ID"])

# Display the data table for preview
complete_trial_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_tumor_volume = complete_trial_df.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].mean()
# Convert to DataFrame
mean_tumor_vol_df = pd.DataFrame(mean_tumor_volume).reset_index()
mean_tumor_vol_df.rename(columns = {"Tumor Volume (mm3)":"Mean Tumor Volume (mm3)"}, inplace = True)
# Preview DataFrame
mean_tumor_vol_df.head()
```

```python
mean_tumor_vol_df.describe()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
stderr_tumor_vol_df = complete_trial_df.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].sem()
# Convert to DataFrame
stderr_tumor_vol_df = pd.DataFrame(stderr_tumor_vol_df).reset_index()
stderr_tumor_vol_df.rename(columns = {"Tumor Volume (mm3)":"Tumor Volume StdErr (mm3)"}, inplace = True)
# Preview DataFrame
stderr_tumor_vol_df.head()
```

```python
#se_tumor_vol_df.index
mean_tumor_vol_df.index
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_tumor_by_time = mean_tumor_vol_df.pivot(index='Timepoint', columns='Drug', values='Mean Tumor Volume (mm3)')
stderr_tumor_by_time = stderr_tumor_vol_df.pivot(index='Timepoint', columns='Drug', values='Tumor Volume StdErr (mm3)')

# Preview that Reformatting worked
mean_tumor_by_time
```

```python
# Preview that Reformatting worked
stderr_tumor_by_time
```

```python
mouse_df["Drug"].unique
```

```python
# Generate the Plot (with Error Bars)
drugs = ["Capomulin", "Infubinol", "Ketapril", "Placebo"]
plotStyleDict = {"Capomulin":'red', "Infubinol":'blue', "Ketapril":'green', "Placebo":'black'}

fig1, ax1 = plt.subplots()
for drug in drugs:
    ax1.errorbar(mean_tumor_by_time.index, mean_tumor_by_time[drug], stderr_tumor_by_time[drug], 
                 fmt = plotStyleDict[drug], label = [drug])
plt.plot()
ax1.legend(loc="best", fontsize="small", fancybox=True)
plt.grid(True)
plt.title('Tumor Response to Treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Tumor Volume (mm3)')

# Save the Figure
plt.savefig("TumorResponseToTreatment.png")
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_met_site = complete_trial_df.groupby(["Drug","Timepoint"])["Metastatic Sites"].mean()
mean_met_site = complete_trial_df.groupby(["Drug","Timepoint"])["Metastatic Sites"].mean()

# Convert to DataFrame
mean_met_site_df = pd.DataFrame(mean_met_site).reset_index()
mean_met_site_df.rename(columns = {"Metastatic Sites":"Mean Metastatic Sites"}, inplace = True)

# Preview DataFrame
mean_met_site_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
stderr_met_site = complete_trial_df.groupby(["Drug","Timepoint"])["Metastatic Sites"].sem()

# Convert to DataFrame
stderr_met_site_df = pd.DataFrame(stderr_met_site).reset_index()
stderr_met_site_df.rename(columns = {"Metastatic Sites":"Metastatic Sites StdErr"}, inplace = True)

# Preview DataFrame
stderr_met_site_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_met_site_by_time = mean_met_site_df.pivot(index='Timepoint', columns='Drug', values='Mean Metastatic Sites')
std_err_met_site_by_time = stderr_met_site_df.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites StdErr')

# Preview that Reformatting worked
mean_met_site_by_time
```

```python
std_err_met_site_by_time
```

```python
# Generate the Plot (with Error Bars)
drugs = ["Capomulin", "Infubinol", "Ketapril", "Placebo"]
plotStyleDict = {"Capomulin":'red', "Infubinol":'blue', "Ketapril":'green', "Placebo":'black'}

fig1, ax1 = plt.subplots()
for drug in drugs:
    ax1.errorbar(mean_met_site_by_time.index, mean_met_site_by_time[drug], std_err_met_site_by_time[drug], 
                 fmt = plotStyleDict[drug], label = [drug])
plt.plot()
ax1.legend(loc="best", fontsize="small", fancybox=True)
plt.grid(True)
plt.title('Metastatic Response to Treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Metastatic Sites')

# Save the Figure
plt.savefig("MetastaticResponseToTreatment.png")
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse_count_df = complete_trial_df.groupby(["Drug","Timepoint"])["Mouse ID"].count()

# Convert to DataFrame
mouse_count_df = pd.DataFrame(mouse_count_df).reset_index()
mouse_count_df.rename(columns = {"Mouse ID":"Mouse Count"}, inplace = True)

# Preview DataFrame
mouse_count_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mouse_count_by_time = mouse_count_df.pivot(index='Timepoint', columns='Drug', values='Mouse Count')

# Preview that Reformatting worked
mouse_count_by_time
```

```python
# Put the counts of mice at each timepoint into a list by drug
capomulin_list = mouse_count_by_time['Capomulin'].tolist()
infubinol_list = mouse_count_by_time['Infubinol'].tolist()
ketapril_list = mouse_count_by_time['Ketapril'].tolist()
placebo_list = mouse_count_by_time['Placebo'].tolist()
```

```python
# Divide each value in the lists by the first value of that list to get a percentage
capomulin_surv_rate = [(x/capomulin_list[0])*100 for x in capomulin_list]
infubinol_surv_rate = [(x/infubinol_list[0])*100 for x in infubinol_list]
ketapril_surv_rate = [(x/ketapril_list[0])*100 for x in ketapril_list]
placebo_surv_rate = [(x/placebo_list[0])*100 for x in placebo_list]
```

```python
# Generate the Plot (Accounting for percentages)

ax = plt.subplot()

# Set the x axis from 0 to 45 in increments of 5
x_axis = np.arange(0, 50, 5)

# Set the plot title and axes titles
plt.title("Survival During Treatment")
plt.xlabel("Time (days)")
plt.ylabel("Survival Rate (%)")

# Plot the 'mean' list vs. the established x axis with error 
ax.plot(x_axis, capomulin_surv_rate, color="red", label="Capomulin")
ax.plot(x_axis, infubinol_surv_rate, color="blue", label="Infubinol")
ax.plot(x_axis, ketapril_surv_rate, color="green", label="Ketapril")
ax.plot(x_axis, placebo_surv_rate, color="black", label="Placebo")

# Add the legend and gridlines
ax.legend(loc='best')

tick_locations = [value for value in x_axis]
ax.set_xticks(tick_locations, minor=False)
ax.grid('on', which='major', axis='both', linestyle='dotted', linewidth=0.5)

plt.xlim(0, max(x_axis))
        
# Show the resulting scatter plot
plt.savefig("SurvivalDuringTreatment.png")
plt.show()
```
